```python
# Another Dataset
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
# Combining Two Datasets
final_df = pd.merge(df, df_country, on='Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
# To Check Data Types
final_df.dtypes
```

```python
# To find with a specific country How many records are there
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Pie Chart
plt.pie(country_val, labels=country_names)
plt.show()
```

```python
## Pie Chart - Top 3 Countries that uses Zomat
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.2f%%")
plt.title("Top 3 Countries using Zomato")
plt.show()
```

**Observation:** Zomato maximum records or transaction are from India after that USA and then United Kingdom.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

**Observation: **

1.   When Rating is between 4.5 to 4.9 ----> Excellent
2.   When Rating is between 4.0 to 3.4 ----> Very good
3.   When Rating is between 3.5 to 3.9 ----> good
4.   When Rating is between 3.0 to 3.4 ----> Average
5.   When Rating is between 2.5 to 4.9 ----> Averge
6.   When Rating is between 2.0 to 4.4 ----> Poor




```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue = 'Rating color', data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue = 'Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```

## Observation:

1.   Not Rated count is very high
2.   Maximum number of rating are between 2.5 to 3.4



```python
sns.countplot(x='Rating color', data = ratings, palette=['blue','red','orange','yellow','green','green'])
```
